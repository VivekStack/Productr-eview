```python
from google.colab import files
import pandas as pd
import io

# Open a file picker dialog
uploaded = files.upload()

# Automatically read the uploaded CSV file
for filename in uploaded.keys():
    df = pd.read_csv(io.BytesIO(uploaded[filename]))
    print(f"Loaded: {filename}")
    print(df.head(5))  # Display first few rows
```

```python
# Basic imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Overview
print("🔍 Dataset Shape:", df.shape)
print("\n📋 First 5 Rows:")
display(df.head())

# 2. Info & Data Types
print("\n📄 Info:")
df.info()

# 3. Missing Values
print("\n❓ Missing Values:")
display(df.isnull().sum())

# 4. Summary Statistics
print("\n📊 Summary Statistics:")
display(df.describe(include='all'))

# 5. Check for Duplicates
print("\n🧬 Duplicated Rows:", df.duplicated().sum())

# 6. Column-wise Unique Values
print("\n🔢 Unique Values per Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")

# 7. Distribution of Numerical Features
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
# Check if numeric_cols is empty
if numeric_cols: # This condition checks if numeric_cols has any elements
  df[numeric_cols].hist(figsize=(14, 10), bins=30, edgecolor='black')
  plt.suptitle("📈 Histograms of Numerical Features")
  plt.show()
else:
  print("No numerical columns found to plot histograms.") # Informative message if no numeric columns

# 8. Count plots for Categorical Features (first 5)
categorical_cols = df.select_dtypes(include='object').columns[:5]
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col)
    plt.xticks(rotation=45)
    plt.title(f"🔢 Count Plot: {col}")
    plt.show()
```

```python
# Data Cleaning
# 1. Check missing values
print("Missing values per column:\n", df.isnull().sum())

# 2. Fill missing values (example: fill with median for numeric)
for col in df.select_dtypes(include='number').columns:
    df[col].fillna(df[col].median(), inplace=True)

# Fill categorical with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# 3. Drop duplicates
df.drop_duplicates(inplace=True)

# 4. Optional: strip whitespace from text columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)


```

```python
#Feature Engineering
# 1. One-Hot Encode categorical features
categorical_cols = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# 2. Scale numeric columns
from sklearn.preprocessing import StandardScaler

# Select numeric columns explicitly, excluding those created during one-hot encoding
numeric_cols = df.select_dtypes(include='number').columns
numeric_cols = [col for col in numeric_cols if col not in categorical_cols]  # Exclude original categorical columns

# Check if there are any numeric columns left to scale
if len(numeric_cols) > 0:
  scaler = StandardScaler()
  df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
else:
  print("No numeric columns found to scale after one-hot encoding.")
```
